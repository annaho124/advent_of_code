## Condition 1

```python
string_filter_out = ['ab', 'cd', 'pq', 'xy']
dict_condition_1 = {
    'String': []
}
```

```python
for row in range(len(input_df)):

    string_to_test = input_df.String[row]
    condition_1 = string_filter_out[0] not in string_to_test
    condition_2 = string_filter_out[1] not in string_to_test
    condition_3 = string_filter_out[2] not in string_to_test
    condition_4 = string_filter_out[3] not in string_to_test

    if condition_1 + condition_2 + condition_3 + condition_4 == 4:

        dict_condition_1['String'].append(string_to_test)
```

```python
df_condition_1 = pd.DataFrame(dict_condition_1)
```

## Condition 2

```python
vowel = ['a', 'e', 'i', 'o', 'u']
dict_condition_2 = {
    'String': []
}
```

```python
for row in range(len(df_condition_1)):

    string_to_test = df_condition_1.String[row]
    
    number = 0
    for letter in string_to_test:

        if letter in vowel:

            number = number + 1

        if number >= 3:

            dict_condition_2['String'].append(string_to_test)
            break
```

```python
df_condition_2 = pd.DataFrame(dict_condition_2)
```

## Condition 3

```python
dict_condition_3 = {
    'String': []
}
```

```python
for row in range(len(df_condition_2)):

    string_to_test = df_condition_2.String[row]

    print(string_to_test)

    for letter in range(len(string_to_test) - 1):

        first_letter = string_to_test[letter: letter + 1]
        second_letter = string_to_test[letter + 1: letter + 2]

        print(first_letter)
        print(second_letter)

        if first_letter == second_letter:

            dict_condition_3['String'].append(string_to_test)
            
            break
```
